```python
#Import Libraries

from bs4 import BeautifulSoup
import pandas as pd
```

```python
#Connect to Website

with open("D:\Order Food Online from India's Best Food Delivery Service _ Swiggy.html",encoding='utf') as html:
    content = html.read()
```

```python
#Extracting the Menu

soup = BeautifulSoup(content,'lxml')
menu = soup.find_all('h3',class_='styles_itemNameText__3ZmZZ')
menu
```

```python
#Using for loop to extract and append it to new list

a=[]
for i in menu:
    e=i.text.strip()
    a.append(e)  

a
```

```python
#Extracting the Price

price = soup.find_all('span',class_='rupee')
price
b=[]
for j in price:
    f=j.text.strip()
    b.append(f)  

b
```

```python
b = b[6:]
b
```

```python
#Extracting the Description


description = soup.find_all('div',class_='styles_itemDesc__3vhM0')
description
c=[]
for k in description:
    g=k.text.strip()
    c.append(g)  

c
```

```python
#Converting list to a dataframe

df_swiggy = pd.DataFrame()
df_swiggy['menu'] = a
df_swiggy
```

```python
df_swiggy['price'] = b
df_swiggy['description'] = c
df_swiggy
```

```python
#Pulling this data to excel

df_swiggy.to_excel("D:\swiggy-dataset.xlsx")
```
